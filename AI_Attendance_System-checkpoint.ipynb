```python
import pandas as pd
from datetime import datetime


students = [
    "Tiyasa",
    "Riya",
    "Aman",
    "Priya",
    "Arjun"
]

attendance = []

print("===================================")
print(" AI ATTENDANCE MANAGEMENT SYSTEM ")
print("===================================")

print("\nType P for Present and A for Absent")


for student in students:

    status = input(f"Is {student} Present or Absent? : ")

    if status.upper() == "P":
        attendance.append("Present")
    else:
        attendance.append("Absent")


date = datetime.now().strftime("%Y-%m-%d")

attendance_data = {
    "Student Name": students,
    "Attendance": attendance,
    "Date": [date] * len(students)
}

attendance_df = pd.DataFrame(attendance_data)

print("\n===== Attendance Recorded Successfully =====")
print(attendance_df)


excel_file = f"attendance_{datetime.now().strftime('%Y%m%d_%H%M%S')}.xlsx"

attendance_df.to_excel(excel_file, index=False)

print("\nAttendance saved successfully!")
print("Excel File Name:", excel_file)


while True:

    print("\n===================================")
    print("      ATTENDANCE QUERY MENU")
    print("===================================")
    print("1. Show Absent Students")
    print("2. Show Present Students")
    print("3. Show Total Students")
    print("4. Show Full Attendance Record")
    print("5. Show Attendance Percentage")
    print("6. Exit")

    choice = input("\nEnter Your Choice: ")

    if choice == "1":

        absent_students = attendance_df[
            attendance_df["Attendance"] == "Absent"
        ]

        print("\nAbsent Students:")

        if len(absent_students) == 0:
            print("No absent students.")
        else:
            print(absent_students[["Student Name"]])

    elif choice == "2":

        present_students = attendance_df[
            attendance_df["Attendance"] == "Present"
        ]

        print("\nPresent Students:")
        print(present_students[["Student Name"]])

    elif choice == "3":

        print("\nTotal Students =", len(students))

    elif choice == "4":

        print("\nFull Attendance Record:")
        print(attendance_df)

    elif choice == "5":

        present_count = len(
            attendance_df[
                attendance_df["Attendance"] == "Present"
            ]
        )

        percentage = (present_count / len(students)) * 100

        print("\nAttendance Percentage =", round(percentage, 2), "%")

    elif choice == "6":

        print("\nThank You For Using The System!")
        break

    else:

        print("\nInvalid Choice! Please Enter 1 to 6.")
```
